```python
import pandas as pd
```

```python
df1=pd.read_csv('1.csv')
df1.head()
```

```python
df2=pd.read_csv('2.csv')
df2.head()
```

```python
df3=pd.read_csv('3.csv')
df3.head()
```

```python
df4=pd.read_csv('4.csv')
df4.head()
```

```python
df5=pd.read_csv('5.csv')
df5.head()
```

```python
merged_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_players.csv', index=False)

print("Merged CSV saved as 'merged_players.csv'")
```
